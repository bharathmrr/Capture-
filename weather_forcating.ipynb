```python
import pandas as pd
```

```python
data=pd.read_csv('/content/weather_forecast_data.csv')
```

```python
data.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
la=LabelEncoder()
sa=StandardScaler()
```

```python
data['Pressure'] = sa.fit_transform(data[['Pressure']])
data['Humidity'] = sa.fit_transform(data[['Humidity']])
data['Wind_Speed'] = sa.fit_transform(data[['Wind_Speed']])
data['Temperature'] = sa.fit_transform(data[['Temperature']])
```

```python
data['Rain']=la.fit_transform(data[['Rain']])
```

```python
data.head(1)
```

```python
from sklearn.model_selection import train_test_split
x=data.drop('Rain',axis=1)
y=data['Rain']
```

```python
from re import X
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
rf=RandomForestClassifier()
dt=DecisionTreeClassifier()
lr=LogisticRegression()
```

```python
rf.fit(X_train,y_train)
```

```python
('training',rf.score(X_train,y_train)*100,('testing',rf.score(X_test,y_test)*100))
```

```python
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
('training',dt.score(X_train,y_train)*100,('testing',dt.score(X_test,y_test)*100))
```

```python
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
('training',lr.score(X_train,y_train)*100,('testing',lr.score(X_test,y_test)*100))
```
