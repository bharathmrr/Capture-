```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['school']=le.fit_transform(data['school'])
data['sex']=le.fit_transform(data['sex'])
data['address']=le.fit_transform(data['address'])
data['famsize']=le.fit_transform(data['famsize'])
data['Pstatus']=le.fit_transform(data['Pstatus'])
data[]
```

```python
data.head(1)
```

```python

data['schoolsup']=le.fit_transform(data['schoolsup'])
data['famsup']=le.fit_transform(data['famsup'])
data['paid']=le.fit_transform(data['paid'])
data['activities']=le.fit_transform(data['activities'])
data['nursery']=le.fit_transform(data['nursery'])
data['higher']=le.fit_transform(data['higher'])
data['internet']=le.fit_transform(data['internet'])
data['romantic']=le.fit_transform(data['romantic'])
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first', sparse_output=False) # sparse=False for numpy array

for column in ['Mjob', 'Fjob', 'reason', 'guardian']:
    # Reshape the column to a 2D array with one column
    encoded_data = ohe.fit_transform(data[[column]])

    # Get feature names and create a DataFrame for the encoded data
    feature_names = ohe.get_feature_names_out([column])
    encoded_df = pd.DataFrame(encoded_data, columns=feature_names, index=data.index)

    # Concatenate the encoded DataFrame with the original DataFrame
    data = pd.concat([data, encoded_df], axis=1)

    # Drop the original categorical column
    data = data.drop(columns=[column])
```

```python
x=data.drop(columns=['G3']).values
y=data['G3'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
(lr.score(x_test,y_test)*100,lr.score(x_train,y_train)*100)
```

```python
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
(dt.score(x_test,y_test)*100 ,dt.score(x_train,y_train)*100)
```

```python
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
```

```python
(rf.score(x_test,y_test)*100,rf.score(x_train,y_train)*100)
```

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(x_train,y_train)
```

```python
(svr.score(x_test,y_test)*100,svr.score(x_train,y_train)*100)
```
