```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('/content/data.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data['country']=data['country'].map({'USA':1})
```

```python
data.drop(columns=['date','statezip','street','city','waterfront','view','condition',],inplace=True)
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
data['price']=s.fit_transform(data[['price']])
data['sqft_living']=s.fit_transform(data[['sqft_living']])
data['sqft_lot']=s.fit_transform(data[['sqft_lot']])
data['sqft_above']=s.fit_transform(data[['sqft_above']])
data['sqft_basement']=s.fit_transform(data[['sqft_basement']])
```

```python
from sklearn.preprocessing import OneHotEncoder
o=OneHotEncoder(drop='first')
```

```python
x=data.drop(columns=['price']).values
y=data['price'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
l=LinearRegression()
s=SVR()
d=DecisionTreeRegressor()
r=RandomForestRegressor()
```

```python
l.fit(x_train,y_train)
s.fit(x_train,y_train)
d.fit(x_train,y_train)
r.fit(x_train,y_train)
```

```python
l.score(x_test,y_test)*100
```
