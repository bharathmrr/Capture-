```python
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
d=load_diabetes()
```

```python
d
```

```python
tar=pd.DataFrame(d.target,columns=['target'])
```

```python
data=pd.DataFrame(d.data,columns=d.feature_names)
```

```python
dat=pd.concat([data,tar],axis=1)
```

```python
dat
```

```python
x=dat.drop('target',axis=1).values
y=dat['target'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
# x_train and x_test are already numpy arrays, so remove .values
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
ypre=model.predict(x_test)
```

```python
r2_score(y_test,ypre)
```
