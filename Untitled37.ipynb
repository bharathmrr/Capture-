```python
da=pd.read_csv('/content/Student_Performance.csv')
```

```python
da.head()
```

```python
x=da['Previous Scores']
y=da['Performance Index']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
min_s=sc.fit_transform(np.array(x_train).reshape(-1,1))
```

```python
fig, (ax,ax1)=plt.subplots(ncols=2,figsize=(10,5))
ax.scatter(x_train,y_train)
ax1.scatter(min_s,y_train)
```

```python
fig, (ax,ax1)=plt.subplots(ncols=2,figsize=(10,5))
import seaborn as sns
sns.distplot(x_train,ax=ax)
sns.distplot(min_s,ax=ax1)
```

** MEAN NORMALIZATION**

```python
from sklearn.preprocessing import normalize
mean_s=normalize(np.array(x_train).reshape(-1,1))
```

```python
fig, (ax,ax1)=plt.subplots(ncols=2,figsize=(10,5))
sns.distplot(x_train,ax=ax)
sns.distplot(mean_s,ax=ax1)
```

**MAXABS SCALING**

```python
from sklearn.preprocessing import MaxAbsScaler
ma=MaxAbsScaler()
max_s=ma.fit_transform(np.array(x_train).reshape(-1,1))
```

```python
fig, (ax,ax1)=plt.subplots(ncols=2,figsize=(10,5))
sns.distplot(x_train,ax=ax)
sns.distplot(max_s,ax=ax1)
```

**robust scalling**

```python
from sklearn.preprocessing import RobustScaler
ro=RobustScaler()
ro_s=ro.fit_transform(np.array(x_train).reshape(-1,1))
```

```python
fig, (ax,ax1)=plt.subplots(ncols=2,figsize=(10,5))
sns.distplot(x_train,ax=ax)
sns.distplot(ro_s,ax=ax1)
```
