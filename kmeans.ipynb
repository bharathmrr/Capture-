```python
import pandas as pd
```

```python
data=pd.read_csv("/content/student_performance (1).csv")
```

```python
data.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
data['Sex']=l.fit_transform(data['Sex'])

data['Performance']=l.fit_transform(data['Performance'])
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
x=data.drop('Performance',axis=1)
y=data['Performance']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3,random_state=42)
kmeans.fit(x_train)
```

```python
kmeans.predict(x_test)
```

```python
import matplotlib.pyplot as plt

# Use 'c' instead of 'hue' and provide the y variable for color mapping
plt.scatter(data['Exam_Score'], data['Study_Time'], c=y)

# Add a colorbar to show the mapping between colors and 'Performance' values
plt.colorbar(label='Performance')

plt.xlabel('Exam_Score')
plt.ylabel('Study_Time')
plt.title('Scatter Plot of Exam Score vs. Study Time')
plt.show()
```
