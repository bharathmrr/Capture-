```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, (a, b) = plt.subplots(1, 2, figsize=(10, 5))

# Use the DataFrame 'data' as the data source for countplot
sns.countplot(x='Operating System', data=data, ax=a)  # Changed data=a to data=data, ax=a
sns.countplot(x='Gender', data=data, ax=b) # Changed data=b to data=data, ax=b

plt.show() # Add this line to display the plot
```

```python
data['Gender'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Gender')
plt.show()
data['Operating System'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```
