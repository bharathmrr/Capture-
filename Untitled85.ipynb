```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("/content/heart.csv")
```

```python
data.head(3)
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data['trestbps']=sc.fit_transform(data[['trestbps']])
data['chol']=sc.fit_transform(data[['chol']])
data['thalach']=sc.fit_transform(data[['thalach']])
```

```python
data.head()
```

```python
x=data.drop('target',axis=1).values
y=data['target'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
(lr.score(x_train,y_train)*100,lr.score(x_test,y_test)*100)
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
```

```python
(svc.score(x_train,y_train)*100,svc.score(x_test,y_test)*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
(dt.score(x_train,y_train)*100,dt.score(x_test,y_test)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
(rf.score(x_train,y_train)*100,rf.score(x_test,y_test)*100)
```

```python
from sklearn.naive_bayes import GaussianNB # Corrected the typo in the module name from 'navie_bayes' to 'naive_bayes'
nb=GaussianNB()
nb.fit(x_train,y_train)
```

```python
(nb.score(x_train,y_train)*100,nb.score(x_test,y_test)*100)
```
