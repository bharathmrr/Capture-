```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/sample_data/california_housing_train.csv")
```

```python
data.head()
```

```python
x=data['total_rooms']
y=data['median_house_value']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train=x_train.values.reshape(-1,1)
y_train=y_train.values.reshape(-1,1)

```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
 # Remove .values.reshape(-1,1) as x_test is already a NumPy array
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=1)
x_train=pca.fit_transform(x_train)
x_test=pca.transform(x_test)
```

```python
plt.scatter(x_train,y_train)
plt.xlabel('total_rooms')
plt.ylabel('median_house_value')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,lr.predict(x_train),color='red')
plt.xlabel('total_rooms')
plt.ylabel('median_house_value')
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_train,lr.predict(x_train))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=4)
x_poly=poly.fit_transform(x_train)
```

```python
from sklearn.linear_model import LinearRegression
lr2=LinearRegression()
lr2.fit(x_poly,y_train)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,lr2.predict(poly.fit_transform(x_train)),color='red')
```

```python
r2_score(y_train,lr2.predict(poly.fit_transform(x_train)))
```
