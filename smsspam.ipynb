```python
import pandas as pd

try:
    data = pd.read_csv('/content/spam.csv', encoding='utf-8')  # First try with utf-8
except UnicodeDecodeError:
    # If utf-8 fails, try with 'latin-1' or 'ISO-8859-1'
    data = pd.read_csv('/content/spam.csv', encoding='latin-1')
    # If none of the above work, you might need to experiment with
```

```python
data.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], inplace=True)
```

```python
data.head(100)
```

```python
import re
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub('', text)
```

```python
for i in range(len(data)):
    data['v2'][i] = remove_urls(data['v2'][i])
```

```python
def lower_case(text):
  return text.lower()
```

```python
for i in range(len(data)):
    data['v2'][i] = lower_case(data['v2'][i])
```

```python
data.head(15)
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab') # Download the missing 'punkt_tab' data
from nltk import word_tokenize
import nltk

from nltk.corpus import stopwords
nltk.download('stopwords')

def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)  # Join the words back into a string


for i in range(len(data)):
    data['v2'][i] = remove_stopwords(data['v2'][i])
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data['v1'] = encoder.fit_transform(data['v1'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data['v2'], data['v1'], test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
model = MultinomialNB()

model.fit(X_train_tfidf, y_train)
y_pred = model.predict(X_test_tfidf)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
p=model.predict(vectorizer.transform(['	Your free ringtone is waiting to be collected']))
if p==1:
  print('spam')
else:
  print('not spam')
```
