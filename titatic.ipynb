**Titanic surival prediction**


```python
import pandas  as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

```

```python
data=sns.load_dataset('titanic')
```

```python
data.head(4)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data['age']=data['age'].fillna(data['age'].mean())
data['deck']=data['deck'].fillna(data['deck'].mode()[0])
data['embark_town']=data['embark_town'].fillna(data['embark_town'].mode()[0])
```

```python
data.isnull().sum()
```

```python
data['embarked']=data['embarked'].fillna(data['embarked'].mode()[0])
```

```python
data.head(4)
```

```python
data['sex']=data['sex'].map({'male':1,'female':0})
data['alone']=data['alone'].map({True:1,False:0})
data['alive']=data['alive'].map({'yes':1,'no':0})
```

```python
data.head(4)
```

```python
data.drop(columns=['class','who','adult_male'],inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
data.head(2)
```

```python
ohe=OneHotEncoder(drop='first')
```

```python
data['embarked']=ohe.fit_transform(data[['embarked']]).toarray()
```

```python
data['deck']=ohe.fit_transform(data[['deck']]).toarray()
```

```python
data['embark_town']=ohe.fit_transform(data[['embark_town']]).toarray()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data['fare']=sc.fit_transform(data[['fare']])
data['age']=sc.fit_transform(data[['age']])
```

```python
x=data.drop(columns=['survived']).values
y=data['survived'].values
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import RidgeClassifier

from sklearn.svm import SVC
rf=RidgeClassifier()
lr=LogisticRegression()
from sklearn.tree import DecisionTreeClassifier

svc=SVC()
dt=DecisionTreeClassifier()
```

**SVC**

```python
svc.fit(x_train,y_train)
```

```python
svc.score(x_test,y_test)*100
```

**RIDGE classfier**

```python
rf.fit(x_train,y_train)
```

```python
rf.score(x_test,y_test)*100
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)*100
```
